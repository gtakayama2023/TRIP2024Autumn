```python
# 検出器とピーク設定

# Ge_Name = "L1"
# ch_1173 = 2820
# ch_1332 = 3200
# sum_width = 7 # 積分範囲のシグマ倍率 

# Ge_Name = "L2"
# ch_1173 = 3260
# ch_1332 = 3700
# sum_width = 5 # 積分範囲のシグマ倍率

# Ge_Name = "L3"
# ch_1173 = 2870
# ch_1332 = 3250
# sum_width = 5 # 積分範囲のシグマ倍率

# Ge_Name = "L4"
# ch_1173 = 3060
# ch_1332 = 3480
# sum_width = 5 # 積分範囲のシグマ倍率

# Ge_Name = "R1"
# ch_1173 = 3010
# ch_1332 = 3420
# sum_width = 5 # 積分範囲のシグマ倍率 

# Ge_Name = "R2"
# ch_1173 = 2850
# ch_1332 = 3250
# sum_width = 5 # 積分範囲のシグマ倍率

# Ge_Name = "R3"
# ch_1173 = 2980
# ch_1332 = 3390
# sum_width = 5 # 積分範囲のシグマ倍率

Ge_Name = "R4"
ch_1173 = 2830
ch_1332 = 3210
sum_width = 5 # 積分範囲のシグマ倍率

draw_range = 200
range_1173_low = ch_1173 - draw_range
range_1173_high = ch_1173 + draw_range
range_1332_low = ch_1332 - draw_range
range_1332_high = ch_1332 + draw_range

# ユーザー設定パラメータ
sum_range_color = r.kViolet  # 積分範囲を示す色

# ヒストグラムの準備
Org = r.gROOT.FindObject(f"{Ge_Name}")
h1173 = Org.Clone(f"{Ge_Name}_1173")
h1173_fit = Org.Clone(f"{Ge_Name}_1173_fit")
h1173.GetXaxis().SetRangeUser(range_1173_low,range_1173_high)
h1173_fit.GetXaxis().SetRangeUser(range_1173_low,range_1173_high)
h1332 = Org.Clone(f"{Ge_Name}_1332")
h1332_fit = Org.Clone(f"{Ge_Name}_1332_fit")
h1332.GetXaxis().SetRangeUser(range_1332_low,range_1332_high)
h1332_fit.GetXaxis().SetRangeUser(range_1332_low,range_1332_high)

# キャンバスの設定
canvas = r.TCanvas(f"{Ge_Name}_reso", f"{Ge_Name}_reso", 700, 700)
canvas.Divide(2,2)
canvas.cd(1)
h1173.Draw()
canvas.cd(2)
h1332.Draw()
canvas.cd(3)
h1173_fit.Draw()
canvas.cd(4)
h1332_fit.Draw()
r.lgy()

# ガウス+直線のフィッティング関数を定義
def gaus_plus_linear(x, par):
    gaus = par[0] * r.TMath.Exp(-0.5 * ((x[0]-par[1])/par[2])**2)
    linear = par[3] + par[4] * x[0]
    return gaus + linear

# フィッティング関数の設定
fit_1173 = r.TF1("fit_1173", gaus_plus_linear, ch_1173 - draw_range/2, ch_1173 + draw_range/2, 5)
fit_1332 = r.TF1("fit_1332", gaus_plus_linear, ch_1332 - draw_range/2, ch_1332 + draw_range/2, 5)

# パラメータの初期値設定
def setup_fit_parameters(hist, fit_func, center):
    max_bin = hist.GetMaximumBin()
    max_value = hist.GetBinContent(max_bin)
    bg_left = hist.GetBinContent(hist.FindBin(center - draw_range/2))
    bg_right = hist.GetBinContent(hist.FindBin(center + draw_range/2))
    bg_avg = (bg_left + bg_right) / 2
    
    fit_func.SetParameter(0, max_value - bg_avg)
    fit_func.SetParameter(1, center)
    fit_func.SetParameter(2, 3.0)
    fit_func.SetParameter(3, bg_avg)
    fit_func.SetParameter(4, 0)

# フィッティングの実行
setup_fit_parameters(h1173_fit, fit_1173, ch_1173)
setup_fit_parameters(h1332_fit, fit_1332, ch_1332)

canvas.cd(3)
h1173_fit.Fit(fit_1173, "R")
canvas.cd(4)
h1332_fit.Fit(fit_1332, "R")

# フィットパラメータの取得
mean_1173 = fit_1173.GetParameter(1)
mean_1332 = fit_1332.GetParameter(1)
sigma_1173 = fit_1173.GetParameter(2)
sigma_1332 = fit_1332.GetParameter(2)

# 積算範囲の設定
sum_range_1173 = (
    int(mean_1173 - sum_width * abs(sigma_1173)),
    int(mean_1173 + sum_width * abs(sigma_1173))
)
sum_range_1332 = (
    int(mean_1332 - sum_width * abs(sigma_1332)),
    int(mean_1332 + sum_width * abs(sigma_1332))
)

# 積算範囲を表示する矩形の作成
def draw_sum_range(hist, range_low, range_high):
    ymax = hist.GetMaximum()
    box = r.TBox(range_low, 0, range_high, ymax)
    box.SetFillStyle(3004)  # より見やすい斜線パターン
    box.SetFillColor(sum_range_color)  # 設定された色を使用
    box.SetLineColor(sum_range_color)
    box.SetFillColorAlpha(sum_range_color, 0.3)  # 透明度を設定
    return box

# 積分範囲を表示する垂直線の作成
def draw_sum_range_lines(hist, range_low, range_high):
    ymax = hist.GetMaximum()
    line_low = r.TLine(range_low, 0, range_low, ymax)
    line_high = r.TLine(range_high, 0, range_high, ymax)
    
    # 線のスタイル設定
    line_low.SetLineColor(sum_range_color)
    line_high.SetLineColor(sum_range_color)
    line_low.SetLineWidth(2)
    line_high.SetLineWidth(2)
    line_low.SetLineStyle(2)  # 破線スタイル
    line_high.SetLineStyle(2)
    
    return line_low, line_high


# BGを描画
bg_1173 = r.TF1("bg_1173", "pol1", sum_range_1173[0], sum_range_1173[1])
bg_1332 = r.TF1("bg_1332", "pol1", sum_range_1332[0], sum_range_1332[1])
bg_1173.SetParameters(fit_1173.GetParameter(3), fit_1173.GetParameter(4))
bg_1332.SetParameters(fit_1332.GetParameter(3), fit_1332.GetParameter(4))

# canvas.cd(3)
# h1173_fit.Draw()
# bg_1173.SetLineColor(r.kRed)
# bg_1173.Draw("same")
# box_1173 = draw_sum_range(h1173_fit, sum_range_1173[0], sum_range_1173[1])
# box_1173.Draw("same")

# canvas.cd(4)
# h1332_fit.Draw()
# bg_1332.SetLineColor(r.kRed)
# bg_1332.Draw("same")
# box_1332 = draw_sum_range(h1332_fit, sum_range_1332[0], sum_range_1332[1])
# box_1332.Draw("same")

# 元のbox.Drawの部分を以下のコードに置き換え
canvas.cd(3)
h1173_fit.Draw()
bg_1173.SetLineColor(r.kRed)
bg_1173.Draw("same")
line_low_1173, line_high_1173 = draw_sum_range_lines(h1173_fit, sum_range_1173[0], sum_range_1173[1])
line_low_1173.Draw("same")
line_high_1173.Draw("same")

canvas.cd(4)
h1332_fit.Draw()
bg_1332.SetLineColor(r.kRed)
bg_1332.Draw("same")
line_low_1332, line_high_1332 = draw_sum_range_lines(h1332_fit, sum_range_1332[0], sum_range_1332[1])
line_low_1332.Draw("same")
line_high_1332.Draw("same")

# canvas.Update()

# エネルギー較正
energy = [1173.2, 1332.5]
a = (energy[1]-energy[0])/(mean_1332 - mean_1173)
b = energy[0] - a * mean_1173
canvas.Draw()

# FWHM計算
FWHM_1173_keV = sigma_1173 * 2 * r.TMath.Sqrt(2 * r.TMath.Log(2)) * a
FWHM_1332_keV = sigma_1332 * 2 * r.TMath.Sqrt(2 * r.TMath.Log(2)) * a

# 積算によるピーク面積計算
def calculate_peak_area(hist, range_low, range_high):
    total_counts = 0
    total_error_sq = 0
    for bin in range(hist.FindBin(range_low), hist.FindBin(range_high) + 1):
        counts = hist.GetBinContent(bin)
        total_counts += counts
        total_error_sq += counts  # カウントの誤差は√N
    return total_counts, r.TMath.Sqrt(total_error_sq)

# BG面積計算
def calculate_bg_area(bg_func, range_low, range_high):
    # 直線BGの積分
    a = bg_func.GetParameter(0)  # 切片
    b = bg_func.GetParameter(1)  # 傾き
    bg_area = a * (range_high - range_low) + b * (range_high**2 - range_low**2) / 2
    return bg_area

# 面積計算
total_1173, error_1173 = calculate_peak_area(h1173_fit, sum_range_1173[0], sum_range_1173[1])
total_1332, error_1332 = calculate_peak_area(h1332_fit, sum_range_1332[0], sum_range_1332[1])

bg_1173_area = calculate_bg_area(bg_1173, sum_range_1173[0], sum_range_1173[1])
bg_1332_area = calculate_bg_area(bg_1332, sum_range_1332[0], sum_range_1332[1])

net_1173 = total_1173 - bg_1173_area
net_1332 = total_1332 - bg_1332_area

# 結果の出力
print("")
print("======================================================")
print(f"積分範囲: {sum_width}σ")
print(f"エネルギー較正式: E = {a:.3f} * x + {b:.3f}")
print(f"1173keV FWHM: {FWHM_1173_keV:.2f} keV")
print(f"1332keV FWHM: {FWHM_1332_keV:.2f} keV")
print("\nピーク解析結果:")
print(f"1173keV (積算範囲: {sum_range_1173[0]} - {sum_range_1173[1]} ch):")
print(f"  総計数: {total_1173:.1f} ± {error_1173:.1f} counts")
print(f"  BG面積: {bg_1173_area:.1f} counts")
print(f"  正味の面積: {net_1173:.1f} ± {error_1173:.1f} counts")
print(f"1332keV (積算範囲: {sum_range_1332[0]} - {sum_range_1332[1]} ch):")
print(f"  総計数: {total_1332:.1f} ± {error_1332:.1f} counts")
print(f"  BG面積: {bg_1332_area:.1f} counts")
print(f"  正味の面積: {net_1332:.1f} ± {error_1332:.1f} counts")
print("======================================================")
```

```python
def analyze_spectrum(histo_name, ch_1173, ch_1332, draw_range=200, sigma_width=3.0):
    """
    Ge検出器のエネルギースペクトルを解析する関数
    
    Parameters:
    -----------
    histo_name : str
        解析対象のヒストグラム名
    ch_1173 : int
        1173keVピークの中心チャンネル概算値
    ch_1332 : int
        1332keVピークの中心チャンネル概算値
    draw_range : int, optional
        表示範囲（中心からの±チャンネル数）, default=200
    sigma_width : float, optional
        積分範囲（シグマの何倍か）, default=3.0
    
    Returns:
    --------
    dict
        解析結果を格納した辞書
    """
    import ROOT as r
    
    # 範囲の設定
    range_1173_low = ch_1173 - draw_range
    range_1173_high = ch_1173 + draw_range
    range_1332_low = ch_1332 - draw_range
    range_1332_high = ch_1332 + draw_range

    # ヒストグラムの準備
    Org = r.gROOT.FindObject(histo_name)
    h1173 = Org.Clone(f"{histo_name}_1173")
    h1173_fit = Org.Clone(f"{histo_name}_1173_fit")
    h1173.GetXaxis().SetRangeUser(range_1173_low, range_1173_high)
    h1173_fit.GetXaxis().SetRangeUser(range_1173_low, range_1173_high)
    h1332 = Org.Clone(f"{histo_name}_1332")
    h1332_fit = Org.Clone(f"{histo_name}_1332_fit")
    h1332.GetXaxis().SetRangeUser(range_1332_low, range_1332_high)
    h1332_fit.GetXaxis().SetRangeUser(range_1332_low, range_1332_high)

    # キャンバスの設定
    canvas = r.TCanvas(f"{histo_name}_reso", f"{histo_name}_reso", 900, 900)
    canvas.Divide(2,2)

    # ガウス+直線のフィッティング関数を定義
    def gaus_plus_linear(x, par):
        gaus = par[0] * r.TMath.Exp(-0.5 * ((x[0]-par[1])/par[2])**2)
        linear = par[3] + par[4] * x[0]
        return gaus + linear

    # フィッティング関数の設定と初期値設定
    def setup_fit(hist, center, range_width):
        fit_func = r.TF1("fit_func", gaus_plus_linear, 
                        center - range_width/2, center + range_width/2, 5)
        max_bin = hist.GetMaximumBin()
        max_value = hist.GetBinContent(max_bin)
        bg_left = hist.GetBinContent(hist.FindBin(center - range_width/2))
        bg_right = hist.GetBinContent(hist.FindBin(center + range_width/2))
        bg_avg = (bg_left + bg_right) / 2
        
        fit_func.SetParameter(0, max_value - bg_avg)
        fit_func.SetParameter(1, center)
        fit_func.SetParameter(2, 3.0)
        fit_func.SetParameter(3, bg_avg)
        fit_func.SetParameter(4, 0)
        return fit_func

    # フィッティングの実行
    fit_1173 = setup_fit(h1173_fit, ch_1173, draw_range)
    fit_1332 = setup_fit(h1332_fit, ch_1332, draw_range)
    
    canvas.cd(3)
    h1173_fit.Fit(fit_1173, "R")
    canvas.cd(4)
    h1332_fit.Fit(fit_1332, "R")

    # フィットパラメータの取得
    mean_1173 = fit_1173.GetParameter(1)
    mean_1332 = fit_1332.GetParameter(1)
    sigma_1173 = fit_1173.GetParameter(2)
    sigma_1332 = fit_1332.GetParameter(2)

    # 積算範囲の設定
    sum_range_1173 = (
        int(mean_1173 - sigma_width * sigma_1173),
        int(mean_1173 + sigma_width * sigma_1173)
    )
    sum_range_1332 = (
        int(mean_1332 - sigma_width * sigma_1332),
        int(mean_1332 + sigma_width * sigma_1332)
    )

    # 積分範囲を表示する矩形の作成
    def draw_sum_range(hist, range_low, range_high):
        ymax = hist.GetMaximum()
        # 半透明の帯を作成
        box = r.TBox(range_low, 0, range_high, ymax)
        box.SetFillStyle(1001)  # 塗りつぶし
        box.SetFillColorAlpha(r.kAzure-9, 0.3)  # 薄い青色で半透明
        box.SetLineColor(r.kAzure+3)  # 濃い青色の枠線
        box.SetLineWidth(2)
        return box

    # BGを描画
    bg_1173 = r.TF1("bg_1173", "pol1", sum_range_1173[0], sum_range_1173[1])
    bg_1332 = r.TF1("bg_1332", "pol1", sum_range_1332[0], sum_range_1332[1])
    bg_1173.SetParameters(fit_1173.GetParameter(3), fit_1173.GetParameter(4))
    bg_1332.SetParameters(fit_1332.GetParameter(3), fit_1332.GetParameter(4))

    # グラフの描画
    canvas.cd(1)
    h1173.Draw()
    canvas.cd(2)
    h1332.Draw()
    
    canvas.cd(3)
    h1173_fit.Draw()
    bg_1173.SetLineColor(r.kRed)
    bg_1173.SetLineWidth(2)
    bg_1173.Draw("same")
    box_1173 = draw_sum_range(h1173_fit, sum_range_1173[0], sum_range_1173[1])
    box_1173.Draw("same")
    h1173_fit.Draw("same")  # ヒストグラムを再度描画して前面に

    canvas.cd(4)
    h1332_fit.Draw()
    bg_1332.SetLineColor(r.kRed)
    bg_1332.SetLineWidth(2)
    bg_1332.Draw("same")
    box_1332 = draw_sum_range(h1332_fit, sum_range_1332[0], sum_range_1332[1])
    box_1332.Draw("same")
    h1332_fit.Draw("same")

    # エネルギー較正
    energy = [1173.2, 1332.5]
    a = (energy[1]-energy[0])/(mean_1332 - mean_1173)
    b = energy[0] - a * mean_1173

    # FWHM計算
    FWHM_1173_keV = sigma_1173 * 2 * r.TMath.Sqrt(2 * r.TMath.Log(2)) * a
    FWHM_1332_keV = sigma_1332 * 2 * r.TMath.Sqrt(2 * r.TMath.Log(2)) * a

    # 面積計算関数
    def calculate_peak_area(hist, range_low, range_high):
        total_counts = 0
        total_error_sq = 0
        for bin in range(hist.FindBin(range_low), hist.FindBin(range_high) + 1):
            counts = hist.GetBinContent(bin)
            total_counts += counts
            total_error_sq += counts
        return total_counts, r.TMath.Sqrt(total_error_sq)

    def calculate_bg_area(bg_func, range_low, range_high):
        a = bg_func.GetParameter(0)
        b = bg_func.GetParameter(1)
        bg_area = a * (range_high - range_low) + b * (range_high**2 - range_low**2) / 2
        return bg_area

    # 面積計算
    total_1173, error_1173 = calculate_peak_area(h1173_fit, sum_range_1173[0], sum_range_1173[1])
    total_1332, error_1332 = calculate_peak_area(h1332_fit, sum_range_1332[0], sum_range_1332[1])

    bg_1173_area = calculate_bg_area(bg_1173, sum_range_1173[0], sum_range_1173[1])
    bg_1332_area = calculate_bg_area(bg_1332, sum_range_1332[0], sum_range_1332[1])

    net_1173 = total_1173 - bg_1173_area
    net_1332 = total_1332 - bg_1332_area

    # 結果の出力
    print("")
    print("======================================================")
    print(f"エネルギー較正式: E = {a:.3f} * x + {b:.3f}")
    print(f"1173keV FWHM: {FWHM_1173_keV:.2f} keV")
    print(f"1332keV FWHM: {FWHM_1332_keV:.2f} keV")
    print("\nピーク解析結果:")
    print(f"積分範囲: ±{sigma_width}σ")
    print(f"1173keV (積算範囲: {sum_range_1173[0]} - {sum_range_1173[1]} ch):")
    print(f"  総計数: {total_1173:.1f} ± {error_1173:.1f} counts")
    print(f"  BG面積: {bg_1173_area:.1f} counts")
    print(f"  正味の面積: {net_1173:.1f} ± {error_1173:.1f} counts")
    print(f"1332keV (積算範囲: {sum_range_1332[0]} - {sum_range_1332[1]} ch):")
    print(f"  総計数: {total_1332:.1f} ± {error_1332:.1f} counts")
    print(f"  BG面積: {bg_1332_area:.1f} counts")
    print(f"  正味の面積: {net_1332:.1f} ± {error_1332:.1f} counts")
    print("======================================================")

    canvas.Draw()
    
    # 結果を辞書として返す
    results = {
        'calibration': {'slope': a, 'intercept': b},
        'FWHM': {'1173': FWHM_1173_keV, '1332': FWHM_1332_keV},
        '1173keV': {
            'range': sum_range_1173,
            'total': total_1173,
            'error': error_1173,
            'background': bg_1173_area,
            'net': net_1173
        },
        '1332keV': {
            'range': sum_range_1332,
            'total': total_1332,
            'error': error_1332,
            'background': bg_1332_area,
            'net': net_1332
        }
    }
    
    return results, canvas

# 使用例
"""
# jupyternotebookでの使用例:
results, canvas = analyze_spectrum(
    histo_name="L1",      # ヒストグラム名
    ch_1173=2800,         # 1173keVピークの概算中心位置
    ch_1332=3180,         # 1332keVピークの概算中心位置
    draw_range=200,       # 表示範囲
    sigma_width=3.0       # 積分範囲（シグマの倍率）
)

# 結果の取得例
print(f"1173keVピークの正味の面積: {results['1173keV']['net']:.1f}")
print(f"エネルギー分解能(1332keV): {results['FWHM']['1332']:.2f} keV")
"""
```

```python
# 検出器とピーク設定
Ge_Name = "L1"
ch_1173 = 2257
ch_1332 = 2564
# ch_1173 = 1173
# ch_1332 = 1332
draw_range = 200
range_1173_low = ch_1173 - draw_range
range_1173_high = ch_1173 + draw_range
range_1332_low = ch_1332 - draw_range
range_1332_high = ch_1332 + draw_range

# ユーザー設定パラメータ
sum_width = 7 # 積分範囲のシグマ倍率 (例: 3σ)
sum_range_color = r.kViolet  # 積分範囲を示す色

# ヒストグラムの準備
Org = r.gROOT.FindObject(f"{Ge_Name}")
h1173 = Org.Clone(f"{Ge_Name}_1173")
h1173_fit = Org.Clone(f"{Ge_Name}_1173_fit")
h1173.GetXaxis().SetRangeUser(range_1173_low,range_1173_high)
h1173_fit.GetXaxis().SetRangeUser(range_1173_low,range_1173_high)
h1332 = Org.Clone(f"{Ge_Name}_1332")
h1332_fit = Org.Clone(f"{Ge_Name}_1332_fit")
h1332.GetXaxis().SetRangeUser(range_1332_low,range_1332_high)
h1332_fit.GetXaxis().SetRangeUser(range_1332_low,range_1332_high)

# キャンバスの設定
canvas = r.TCanvas(f"{Ge_Name}_reso", f"{Ge_Name}_reso", 700, 700)
canvas.Divide(2,2)
canvas.cd(1)
h1173.Draw()
canvas.cd(2)
h1332.Draw()
canvas.cd(3)
h1173_fit.Draw()
canvas.cd(4)
h1332_fit.Draw()
r.lgy()

# ガウス+直線のフィッティング関数を定義
def gaus_plus_linear(x, par):
    gaus = par[0] * r.TMath.Exp(-0.5 * ((x[0]-par[1])/par[2])**2)
    linear = par[3] + par[4] * x[0]
    return gaus + linear

# フィッティング関数の設定
fit_1173 = r.TF1("fit_1173", gaus_plus_linear, ch_1173 - draw_range/2, ch_1173 + draw_range/2, 5)
fit_1332 = r.TF1("fit_1332", gaus_plus_linear, ch_1332 - draw_range/2, ch_1332 + draw_range/2, 5)

# パラメータの初期値設定
def setup_fit_parameters(hist, fit_func, center):
    max_bin = hist.GetMaximumBin()
    max_value = hist.GetBinContent(max_bin)
    bg_left = hist.GetBinContent(hist.FindBin(center - draw_range/2))
    bg_right = hist.GetBinContent(hist.FindBin(center + draw_range/2))
    bg_avg = (bg_left + bg_right) / 2
    
    fit_func.SetParameter(0, max_value - bg_avg)
    fit_func.SetParameter(1, center)
    fit_func.SetParameter(2, 3.0)
    fit_func.SetParameter(3, bg_avg)
    fit_func.SetParameter(4, 0)

# フィッティングの実行
setup_fit_parameters(h1173_fit, fit_1173, ch_1173)
setup_fit_parameters(h1332_fit, fit_1332, ch_1332)

canvas.cd(3)
h1173_fit.Fit(fit_1173, "R")
canvas.cd(4)
h1332_fit.Fit(fit_1332, "R")

# フィットパラメータの取得
mean_1173 = fit_1173.GetParameter(1)
mean_1332 = fit_1332.GetParameter(1)
sigma_1173 = fit_1173.GetParameter(2)
sigma_1332 = fit_1332.GetParameter(2)

# 積算範囲の設定
sum_range_1173 = (
    int(mean_1173 - sum_width * sigma_1173),
    int(mean_1173 + sum_width * sigma_1173)
)
sum_range_1332 = (
    int(mean_1332 - sum_width * sigma_1332),
    int(mean_1332 + sum_width * sigma_1332)
)

# 積算範囲を表示する矩形の作成
def draw_sum_range(hist, range_low, range_high):
    ymax = hist.GetMaximum()
    box = r.TBox(range_low, 0, range_high, ymax)
    box.SetFillStyle(3004)  # より見やすい斜線パターン
    box.SetFillColor(sum_range_color)  # 設定された色を使用
    box.SetLineColor(sum_range_color)
    box.SetFillColorAlpha(sum_range_color, 0.3)  # 透明度を設定
    return box

# BGを描画
bg_1173 = r.TF1("bg_1173", "pol1", sum_range_1173[0], sum_range_1173[1])
bg_1332 = r.TF1("bg_1332", "pol1", sum_range_1332[0], sum_range_1332[1])
bg_1173.SetParameters(fit_1173.GetParameter(3), fit_1173.GetParameter(4))
bg_1332.SetParameters(fit_1332.GetParameter(3), fit_1332.GetParameter(4))

canvas.cd(3)
h1173_fit.Draw()
bg_1173.SetLineColor(r.kRed)
bg_1173.Draw("same")
box_1173 = draw_sum_range(h1173_fit, sum_range_1173[0], sum_range_1173[1])
box_1173.Draw("same")

canvas.cd(4)
h1332_fit.Draw()
bg_1332.SetLineColor(r.kRed)
bg_1332.Draw("same")
box_1332 = draw_sum_range(h1332_fit, sum_range_1332[0], sum_range_1332[1])
box_1332.Draw("same")

# エネルギー較正
energy = [1173.2, 1332.5]
a = (energy[1]-energy[0])/(mean_1332 - mean_1173)
b = energy[0] - a * mean_1173
canvas.Draw()

# FWHM計算
FWHM_1173_keV = sigma_1173 * 2 * r.TMath.Sqrt(2 * r.TMath.Log(2)) * a
FWHM_1332_keV = sigma_1332 * 2 * r.TMath.Sqrt(2 * r.TMath.Log(2)) * a

# 積算によるピーク面積計算
def calculate_peak_area(hist, range_low, range_high):
    total_counts = 0
    total_error_sq = 0
    for bin in range(hist.FindBin(range_low), hist.FindBin(range_high) + 1):
        counts = hist.GetBinContent(bin)
        total_counts += counts
        total_error_sq += counts  # カウントの誤差は√N
    return total_counts, r.TMath.Sqrt(total_error_sq)

# BG面積計算
def calculate_bg_area(bg_func, range_low, range_high):
    # 直線BGの積分
    a = bg_func.GetParameter(0)  # 切片
    b = bg_func.GetParameter(1)  # 傾き
    bg_area = a * (range_high - range_low) + b * (range_high**2 - range_low**2) / 2
    return bg_area

# 面積計算
total_1173, error_1173 = calculate_peak_area(h1173_fit, sum_range_1173[0], sum_range_1173[1])
total_1332, error_1332 = calculate_peak_area(h1332_fit, sum_range_1332[0], sum_range_1332[1])

bg_1173_area = calculate_bg_area(bg_1173, sum_range_1173[0], sum_range_1173[1])
bg_1332_area = calculate_bg_area(bg_1332, sum_range_1332[0], sum_range_1332[1])

net_1173 = total_1173 - bg_1173_area
net_1332 = total_1332 - bg_1332_area

# 結果の出力
print("")
print("======================================================")
print(f"積分範囲: {sum_width}σ")
print(f"エネルギー較正式: E = {a:.3f} * x + {b:.3f}")
print(f"1173keV FWHM: {FWHM_1173_keV:.2f} keV")
print(f"1332keV FWHM: {FWHM_1332_keV:.2f} keV")
print("\nピーク解析結果:")
print(f"1173keV (積算範囲: {sum_range_1173[0]} - {sum_range_1173[1]} ch):")
print(f"  総計数: {total_1173:.1f} ± {error_1173:.1f} counts")
print(f"  BG面積: {bg_1173_area:.1f} counts")
print(f"  正味の面積: {net_1173:.1f} ± {error_1173:.1f} counts")
print(f"1332keV (積算範囲: {sum_range_1332[0]} - {sum_range_1332[1]} ch):")
print(f"  総計数: {total_1332:.1f} ± {error_1332:.1f} counts")
print(f"  BG面積: {bg_1332_area:.1f} counts")
print(f"  正味の面積: {net_1332:.1f} ± {error_1332:.1f} counts")
print("======================================================")
```

```python
# 検出器とピーク設定
Ge_Name = "L1"
ch_1173 = 2420
ch_1332 = 2750
draw_range = 100
range_1173_low = ch_1173 - draw_range
range_1173_high = ch_1173 + draw_range
range_1332_low = ch_1332 - draw_range
range_1332_high = ch_1332 + draw_range

# ユーザー設定パラメータ
sum_width = 7 # 積分範囲のシグマ倍率 (例: 3σ)
sum_range_color = r.kViolet  # 積分範囲を示す色

# ヒストグラムの準備
Org = r.gROOT.FindObject(f"{Ge_Name}")
h1173 = Org.Clone(f"{Ge_Name}_1173")
h1173_fit = Org.Clone(f"{Ge_Name}_1173_fit")
h1173.GetXaxis().SetRangeUser(range_1173_low,range_1173_high)
h1173_fit.GetXaxis().SetRangeUser(range_1173_low,range_1173_high)
h1332 = Org.Clone(f"{Ge_Name}_1332")
h1332_fit = Org.Clone(f"{Ge_Name}_1332_fit")
h1332.GetXaxis().SetRangeUser(range_1332_low,range_1332_high)
h1332_fit.GetXaxis().SetRangeUser(range_1332_low,range_1332_high)

# キャンバスの設定
canvas = r.TCanvas(f"{Ge_Name}_reso", f"{Ge_Name}_reso", 700, 700)
canvas.Divide(2,2)
canvas.cd(1)
h1173.Draw()
canvas.cd(2)
h1332.Draw()
canvas.cd(3)
h1173_fit.Draw()
canvas.cd(4)
h1332_fit.Draw()
r.lgy()

# ガウス+直線のフィッティング関数を定義
def gaus_plus_linear(x, par):
    gaus = par[0] * r.TMath.Exp(-0.5 * ((x[0]-par[1])/par[2])**2)
    linear = par[3] + par[4] * x[0]
    return gaus + linear

# フィッティング関数の設定
fit_1173 = r.TF1("fit_1173", gaus_plus_linear, ch_1173 - draw_range/2, ch_1173 + draw_range/2, 5)
fit_1332 = r.TF1("fit_1332", gaus_plus_linear, ch_1332 - draw_range/2, ch_1332 + draw_range/2, 5)

# パラメータの初期値設定
def setup_fit_parameters(hist, fit_func, center):
    max_bin = hist.GetMaximumBin()
    max_value = hist.GetBinContent(max_bin)
    bg_left = hist.GetBinContent(hist.FindBin(center - draw_range/2))
    bg_right = hist.GetBinContent(hist.FindBin(center + draw_range/2))
    bg_avg = (bg_left + bg_right) / 2
    
    fit_func.SetParameter(0, max_value - bg_avg)
    fit_func.SetParameter(1, center)
    fit_func.SetParameter(2, 3.0)
    fit_func.SetParameter(3, bg_avg)
    fit_func.SetParameter(4, 0)

# フィッティングの実行
setup_fit_parameters(h1173_fit, fit_1173, ch_1173)
setup_fit_parameters(h1332_fit, fit_1332, ch_1332)

canvas.cd(3)
h1173_fit.Fit(fit_1173, "R")
canvas.cd(4)
h1332_fit.Fit(fit_1332, "R")

# フィットパラメータの取得
mean_1173 = fit_1173.GetParameter(1)
mean_1332 = fit_1332.GetParameter(1)
sigma_1173 = fit_1173.GetParameter(2)
sigma_1332 = fit_1332.GetParameter(2)

# 積算範囲の設定
sum_range_1173 = (
    int(mean_1173 - sum_width * sigma_1173),
    int(mean_1173 + sum_width * sigma_1173)
)
sum_range_1332 = (
    int(mean_1332 - sum_width * sigma_1332),
    int(mean_1332 + sum_width * sigma_1332)
)

# 積算範囲を表示する矩形の作成
def draw_sum_range(hist, range_low, range_high):
    ymax = hist.GetMaximum()
    box = r.TBox(range_low, 0, range_high, ymax)
    box.SetFillStyle(3004)  # より見やすい斜線パターン
    box.SetFillColor(sum_range_color)  # 設定された色を使用
    box.SetLineColor(sum_range_color)
    box.SetFillColorAlpha(sum_range_color, 0.3)  # 透明度を設定
    return box

# BGを描画
bg_1173 = r.TF1("bg_1173", "pol1", sum_range_1173[0], sum_range_1173[1])
bg_1332 = r.TF1("bg_1332", "pol1", sum_range_1332[0], sum_range_1332[1])
bg_1173.SetParameters(fit_1173.GetParameter(3), fit_1173.GetParameter(4))
bg_1332.SetParameters(fit_1332.GetParameter(3), fit_1332.GetParameter(4))

canvas.cd(3)
h1173_fit.Draw()
bg_1173.SetLineColor(r.kRed)
bg_1173.Draw("same")
box_1173 = draw_sum_range(h1173_fit, sum_range_1173[0], sum_range_1173[1])
box_1173.Draw("same")

canvas.cd(4)
h1332_fit.Draw()
bg_1332.SetLineColor(r.kRed)
bg_1332.Draw("same")
box_1332 = draw_sum_range(h1332_fit, sum_range_1332[0], sum_range_1332[1])
box_1332.Draw("same")

# エネルギー較正
energy = [1173.2, 1332.5]
a = (energy[1]-energy[0])/(mean_1332 - mean_1173)
b = energy[0] - a * mean_1173
canvas.Draw()

# FWHM計算
FWHM_1173_keV = sigma_1173 * 2 * r.TMath.Sqrt(2 * r.TMath.Log(2)) * a
FWHM_1332_keV = sigma_1332 * 2 * r.TMath.Sqrt(2 * r.TMath.Log(2)) * a

# 積算によるピーク面積計算
def calculate_peak_area(hist, range_low, range_high):
    total_counts = 0
    total_error_sq = 0
    for bin in range(hist.FindBin(range_low), hist.FindBin(range_high) + 1):
        counts = hist.GetBinContent(bin)
        total_counts += counts
        total_error_sq += counts  # カウントの誤差は√N
    return total_counts, r.TMath.Sqrt(total_error_sq)

# BG面積計算
def calculate_bg_area(bg_func, range_low, range_high):
    # 直線BGの積分
    a = bg_func.GetParameter(0)  # 切片
    b = bg_func.GetParameter(1)  # 傾き
    bg_area = a * (range_high - range_low) + b * (range_high**2 - range_low**2) / 2
    return bg_area

# 面積計算
total_1173, error_1173 = calculate_peak_area(h1173_fit, sum_range_1173[0], sum_range_1173[1])
total_1332, error_1332 = calculate_peak_area(h1332_fit, sum_range_1332[0], sum_range_1332[1])

bg_1173_area = calculate_bg_area(bg_1173, sum_range_1173[0], sum_range_1173[1])
bg_1332_area = calculate_bg_area(bg_1332, sum_range_1332[0], sum_range_1332[1])

net_1173 = total_1173 - bg_1173_area
net_1332 = total_1332 - bg_1332_area

# 結果の出力
print("")
print("======================================================")
print(f"積分範囲: {sum_width}σ")
print(f"エネルギー較正式: E = {a:.3f} * x + {b:.3f}")
print(f"1173keV FWHM: {FWHM_1173_keV:.2f} keV")
print(f"1332keV FWHM: {FWHM_1332_keV:.2f} keV")
print("\nピーク解析結果:")
print(f"1173keV (積算範囲: {sum_range_1173[0]} - {sum_range_1173[1]} ch):")
print(f"  総計数: {total_1173:.1f} ± {error_1173:.1f} counts")
print(f"  BG面積: {bg_1173_area:.1f} counts")
print(f"  正味の面積: {net_1173:.1f} ± {error_1173:.1f} counts")
print(f"1332keV (積算範囲: {sum_range_1332[0]} - {sum_range_1332[1]} ch):")
print(f"  総計数: {total_1332:.1f} ± {error_1332:.1f} counts")
print(f"  BG面積: {bg_1332_area:.1f} counts")
print(f"  正味の面積: {net_1332:.1f} ± {error_1332:.1f} counts")
print("======================================================")
```
